```python
# add random entries

import random
import requests
path = "/api/events/"
base_url = "http://localhost:8002"
create_endpoint = f"{base_url}{path}"
event_count = 10_000
pages = ["/about", "/contact", "/pages", "pricing"]
for i in range(event_count):
    page = random.choice(pages)
    response = requests.post(create_endpoint, json={"page": page})
    if response.ok:
        data = response.json()
        print(data)
    else:
        print(response.text)
```
